```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

Plotting a vector

```python
plt.quiver(0,0,4,5)
plt.show()   #quiver is the fuction that represents a vector
```

```python
plt.quiver(0,0,3,2, scale_units = 'xy', angles = 'xy', scale = 1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,3,2, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,3,2, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,-1,-6, scale_units = 'xy', angles = 'xy', scale = 1, color = 'r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Addition of two vectors


```python
vector_1 = np.asarray([0,0,3,5])

vector_2 = np.asarray([0,0,4,-6])

sum = vector_1 + vector_2
print(sum)
```

```python
plt.quiver(0,0,3,5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,4,-6, scale_units = 'xy', angles = 'xy', scale = 1, color = 'g')
plt.quiver(0,0,7,-1, scale_units = 'xy', angles = 'xy', scale = 1, color = 'r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Subtraction of two Vectors


```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])

difference = vector_1 - vector_2
print(difference)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,3,-2, scale_units = 'xy', angles = 'xy', scale = 1, color = 'g')
plt.quiver(0,0,-1,5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Multiplying a vector by a scalar

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
vector_1 = np.asarray([0,0,2,3])
vector_2 = 2*vector_1
print(vector_2)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,4,6, scale_units = 'xy', angles = 'xy', scale = 1, color = 'g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = -0.5*vector_1
print(vector_2)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,-1,-1.5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
